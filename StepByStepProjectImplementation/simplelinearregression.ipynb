```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('height-weight.csv')
df.head()
```

```python
plt.scatter(df['Height'],df['Weight'])
```

```python
X = df[['Weight']]
Y = df['Height']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.scatter(X_train_scaled,Y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression()
linReg.fit(X_train_scaled,Y_train)
```

```python
linReg.coef_
```

```python
linReg.intercept_
```

```python
y_pred = linReg.predict(X_test_scaled)
```

```python
Y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score,mean_squared_error 
score = r2_score(Y_test,y_pred)
print("Score is ",score)
print("the mse is ",mean_squared_error(Y_test,y_pred))
```

```python
plt.scatter(X_train_scaled,Y_train)
plt.plot(X_train_scaled,linReg.predict(X_train_scaled),'r')
```

```python
plt.scatter(X_test_scaled,Y_test)
plt.plot(X_test_scaled,linReg.predict(X_test_scaled))
```

```python
from sklearn.linear_model import Ridge
rg = Ridge(alpha=1)
```

```python
rg.fit(X_train_scaled,Y_train)
```

```python
pred = rg.predict(X_test_scaled)
from sklearn.metrics import r2_score,mean_squared_error
score = r2_score(Y_test,pred)
print(score)
print(mean_squared_error(Y_test,pred))
```

```python
plt.scatter(X_train_scaled,Y_train)
plt.plot(X_train_scaled,rg.predict(X_train_scaled))
```

```python
plt.scatter(X_test_scaled,Y_test)
plt.plot(X_test_scaled,rg.predict(X_test_scaled))
```
